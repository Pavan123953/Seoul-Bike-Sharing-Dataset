```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pickle
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv("Group_5_data_cleaned.csv")
df.head()
```

```python
df = df.drop(['Unnamed: 0'],axis=1)
```

```python
df.columns
```

```python
import matplotlib.pyplot as plt

from sklearn import preprocessing
from sklearn.decomposition import PCA

data_scaled = preprocessing.scale(df)
df_scaled = pd.DataFrame(data_scaled, columns = df.columns)  

df_scaled  
```

```python
pca = PCA()
X_pca = pca.fit_transform(df_scaled)

# Convert to dataframe
component_names = [f"PC{i+1}" for i in range(X_pca.shape[1])]
X_pca = pd.DataFrame(X_pca, columns=component_names)

X_pca.head()
```

```python
pca.get_covariance()
```

```python
explained_variance=pca.explained_variance_ratio_
explained_variance
```

```python
loadings = pd.DataFrame(
    pca.components_.T,  # transpose the matrix of loadings
    columns=component_names,  # so the columns are the principal components
    index=df_scaled.columns,  # and the rows are the original features
)
loadings
```

```python
with plt.style.context('seaborn-white'):
    plt.figure(figsize=(10, 8))

    plt.bar(range(18), explained_variance, alpha=0.5, align='center',
            label='individual explained variance',color='red')
    plt.ylabel('Explained variance ratio')
    plt.xlabel('Principal components')
    plt.legend(loc='best')
    plt.tight_layout()
```

# first 3 principal components are explaining about the 50% of the data
# so selecting them for the visualization

```python
pca_df = pd.DataFrame(X_pca)
```

```python
pca_df.head(2)
```

```python
# Taking first 2 components for visualization
plt.figure(figsize=(15,8))

plt.scatter(pca_df.PC1, pca_df.PC2, c = df["Rented Bike Count"])
plt.xlabel("PC1", fontsize = 15)
plt.ylabel("PC2", fontsize = 15)
plt.title("Two-component PCA", fontsize = 15)
plt.grid();
```

```python
# Taking first and last components for visualization
plt.figure(figsize=(15,8))

plt.scatter(pca_df.PC1, pca_df.PC3, c = df["Rented Bike Count"])
plt.xlabel("PC1", fontsize = 15)
plt.ylabel("PC3", fontsize = 15)
plt.title("Two-component PCA", fontsize = 15)
plt.grid();
```

```python
# Taking last 2 components for visualization
plt.figure(figsize=(15,8))

plt.scatter(pca_df.PC2, pca_df.PC3, c = df["Rented Bike Count"])
plt.xlabel("PC2", fontsize = 15)
plt.ylabel("PC3", fontsize = 15)
plt.title("Two-component PCA", fontsize = 15)
plt.grid();
```

# Non Linear Dimensionality reduction - kernel PCA

```python
from sklearn.decomposition import KernelPCA
kpca= KernelPCA(n_components=None,kernel = 'rbf',gamma = 10)
```

```python
df_kernel = kpca.fit_transform(df_scaled[1:10000])
```

```python
new_df = df_scaled[1:10000]
k = new_df['Rented Bike Count']
```

```python
plt.figure(figsize=(15,8))

plt.scatter(df_kernel[:,0], df_kernel[:,1] ,c = k)
plt.xlabel("PC1", fontsize = 15)
plt.ylabel("PC2", fontsize = 15)
plt.title("2 component kernel PCA", fontsize = 15)
plt.grid();
```
